```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix

%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')

recent_grads.head(3)
```

```python
recent_grads.iloc[0]
```

```python
recent_grads = recent_grads.dropna()
```

```python
recent_grads.iloc[:,2:].describe()
```

```python
recent_grads.plot(x='ShareWomen',
                  y='Unemployment_rate',
                  kind='scatter',
                  title='ShareWomen vs. Unemployment_rate',
                  figsize=(5,5))
```

```python
x_s = ['Sample_size', 'Sample_size', 'Full_time', 'ShareWomen', 'Men', 'Women']
y_s = ['Median', 'Unemployment_rate', 'Median', 'Unemployment_rate', 'Median', 'Median']

fig = plt.figure(figsize=(17,10))

for i in range(6):
    plt.subplot(2,3, i+1)
    plt.scatter(recent_grads[x_s[i]], recent_grads[y_s[i]])
    plt.xlabel(x_s[i])
    plt.ylabel(y_s[i])
    
plt.title('Relations among some variables')
```

```python
recent_grads.Major_category.value_counts()
```

```python
to_hist = ['Sample_size','Median', 'Employed',
           'Full_time', 'ShareWomen', 'Unemployment_rate',
           'Men', 'Women']

fig = plt.figure(figsize=(10,10))

for i in range(8):
    plt.subplot(3,3, i+1)
    plt.hist(recent_grads[to_hist[i]], bins = 20)
    plt.xlabel(to_hist[i])
```

```python
recent_grads['Median'].hist(bins=25)
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']],
               figsize=(8,8),
               diagonal='kde')
```

```python
recent_grads.sort_values(by = ['Median'],ascending=False)[:10].plot.bar(x='Major', y = 'ShareWomen')
```

```python
recent_grads.sort_values(by = ['Median'],ascending=True)[:10].plot.bar(x='Major', y = 'ShareWomen')
```

```python
recent_grads.sort_values(by = ['Median'],ascending=False)[:10].plot.bar(x='Major', y = 'Unemployment_rate')
recent_grads.sort_values(by = ['Median'],ascending=True)[:10].plot.bar(x='Major', y = 'Unemployment_rate')
```
